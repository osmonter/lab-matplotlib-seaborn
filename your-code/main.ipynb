#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

fig, ax1 = plt.subplots() 

ax1.plot(x, x**2, color = 'blue')
plt.yscale('linear') # Escala lineal para y = x**2

ax2 = ax1.twinx() # Duplicamos eje X pero creamos un nuevo eje Y para la segunda gráfica.

ax2.plot(x, np.exp(x), color = 'red')
plt.yscale('log') # Escala logarítmica para y = np.exp(x)

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Definimos los ejes, una fila con dos gráficas, y el tamaño.

fig, [ax1, ax2] = plt.subplots(1, 2, figsize = (12,4))

# MEDIAN STEPS BY MONTH_ENCODED
# your code here

steps_bymonth = fitbit.groupby(fitbit["Months_encoded"])["Steps"].agg("median")
plt.sca(ax1) # Para que salga como primer gráfico
steps_bymonth.plot(kind = 'line')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

steps_work_or_weekend = fitbit.groupby(fitbit["Work_or_Weekend"])["Steps"].agg("median")
plt.sca(ax2) # Para que salga como segundo gráfico
steps_work_or_weekend.plot(kind = 'bar')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

features = ['Minutes Lightly Active', 'Minutes Very Active','Minutes Sedentary']

for i in features:
    plt.scatter(fitbit[i], fitbit['Steps'])
    plt.show()
```
